```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_s_triangle(n=100):
    top = (np.random.randint(51), np.random.randint(51))
    bottom_left = (np.random.randint(0, top[0]), np.random.randint(0, top[1]))
    bottom_right = (top[0] + top[0] - bottom_left[0], bottom_left[1])
    initial_starting_point = (np.random.randint(bottom_left[0], bottom_right[0]), np.random.randint(bottom_left[1], top[1]))

    point_number_mapper = {1: top,
                           2: bottom_left,
                           3: bottom_right}

    plt.figure(figsize=(10,6))
    plt.plot(top[0], top[1], '.', c='black')
    plt.plot(bottom_left[0], bottom_left[1], '.', c='black')
    plt.plot(bottom_right[0], bottom_right[1], '.', c='black')
    plt.plot(initial_starting_point[0], initial_starting_point[1], '.', c='black')

    points = [initial_starting_point]
    for i in range(n):
        ref_point = np.random.randint(1,4)
        new_point = (((points[-1])[0] + point_number_mapper[ref_point][0] ) / 2,
                     ((points[-1][1] + point_number_mapper[ref_point][1]) / 2))
        points.append(new_point)
        plt.plot(new_point[0], new_point[1], '.', c='black')   
    plt.show()
```

```python
plot_s_triangle()
```

```python
plot_s_triangle(1000)
```
